```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.common.exceptions import TimeoutException
from selenium.webdriver.common.keys import Keys
from bs4 import BeautifulSoup
import time
import pandas as pd

class ScrapeLazada():
    def scrape(self, num_pages=102):
        base_url = 'https://www.lazada.co.th/tag/%E0%B8%9C%E0%B9%89%E0%B8%B2%E0%B8%AD%E0%B8%99%E0%B8%B2%E0%B8%A1%E0%B8%B1%E0%B8%A2-sofy/?page=102&q=%E0%B8%9C%E0%B9%89%E0%B8%B2%E0%B8%AD%E0%B8%99%E0%B8%B2%E0%B8%A1%E0%B8%B1%E0%B8%A2%20sofy&catalog_redirect_tag=true'

        driver = webdriver.Chrome()

        products = []

        for page in range(1, num_pages + 1):
            url = base_url.format(page)
            driver.get(url)

            WebDriverWait(driver, 5).until(EC.presence_of_element_located((By.CSS_SELECTOR, "#root")))
            time.sleep(2)

            soup = BeautifulSoup(driver.page_source, "html.parser")
            for item in soup.find_all('div', class_='Bm3ON'):
                product_name = item.find('div', class_='RfADt').text
                price = item.find('span', class_='ooOxS').text
                sales_element = item.find('span', class_='_1cEkb')
                sales = sales_element.text if sales_element else 'N/A'
                location = item.find('span', class_='oa6ri').text
                products.append((product_name, price, sales, location))

        # Corrected the DataFrame creation
        df = pd.DataFrame(products, columns=['product_name', 'price', 'sales', 'location'])

        # Display DataFrame in Jupyter Notebook or IPython
        from IPython.display import display
        display(df)

        driver.close()

        # Save DataFrame to Excel
        df.to_excel('sofy.xlsx', index=False)
        print('Data saved to sofy.xlsx')

# Instantiate the class and run the scraping
sl = ScrapeLazada()
sl.scrape(num_pages=102)
```
